Anwsers: The data was downloaded from the website of California office of environmental health harzard assessment. It measures the impact of air pollution specific to more than 1000 individual communities throughout the state of California that differ in terms of social and economic conditions. It contains socioeconomic variables such as unemployment rate and high education rate; environmental variables including the concentration and density of PM2.5 and Ozone in each surveyed community; as well as variables that reflect underlying health conditions in each community, including the rates of Ashma and hearth disease.

Using this dataset can measure the correlations between community residents’ socioeconomic vulnerability, health vulnerability (to diseases such as Covid-19), and exposure to air pollution. This aims to imply that communities suffering socioeconomic disadvantages also have high exposure to severe environmental issues (especially air pollution) in associated with underlying health conditions, posing vulnerability to Covid-19 infections. This suggests community environment is an essential determinant of the health crisis.
