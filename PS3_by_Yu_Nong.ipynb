First project: exploring the way child marriage affect children and youth's completion of education, as well as correlaton of level of child and youth education, the frequencies of violence/terrorist/extremist activties, and the use of child labor.

```python
Marriage=pd.read_excel('https://github.com/shimjide123/datman/raw/main/UNICEF%20Child-marriage-database.xlsx')
```

```python
Marriage.head()
```

```python
Marriage.rename(columns={'Child marriage (%)': 'Female_married_by_15(%)'}, inplace=True)
Marriage.rename(columns={'Unnamed: 2': 'Female_married_by_18(%)'}, inplace=True)
Marriage.rename(columns={'Unnamed: 3': 'Male_married_by_18(%)'}, inplace=True)
Marriage.rename(columns={'Countries and areas': 'Countries'}, inplace=True)
```

```python
Marriage['Female_married_by_18(%)'] = pd.to_numeric(Marriage['Female_married_by_18(%)'], errors='coerce')
Marriage['Female_married_by_15(%)'] = pd.to_numeric(Marriage['Female_married_by_15(%)'], errors='coerce')
Marriage['Male_married_by_18(%)'] = pd.to_numeric(Marriage['Male_married_by_18(%)'], errors='coerce')
```

```python
Marriage.dtypes
```

```python
Marriage=Marriage.drop(Marriage.index[[0,1,2]],axis=0)
```

```python
Marriage.reset_index(drop=True)
Marriage.head()
```

```python
Marriage=Marriage.fillna('') #a command of cleaning up the "NaN" values
Marriage.head()
```

```python
for x in Marriage.index: #a command of dropping empty values under one variable column if you only wanna keep the actual values
 if Marriage.loc[x,"Female_married_by_15(%)"]=='':
    Marriage.drop(x,inplace=True)

Marriage.head()
```

```python
for x in Marriage.index:
 if Marriage.loc[x,"Female_married_by_18(%)"]=='':
    Marriage.drop(x,inplace=True)
```
