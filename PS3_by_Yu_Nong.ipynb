```python
import pandas as pd
import pandas_datareader as pdr
from pandas_datareader import wb
from pandas.io.formats.style import Styler
#s4 = Styler(df4, uuid_len=0, cell_ids=False)

from google.colab import files
```

```python
import numpy as np
```

```python
Marriage=pd.read_excel('https://github.com/shimjide123/datman/raw/main/UNICEF%20Child-marriage-database.xlsx')
```

```python
Marriage.head()
```

```python
Marriage.rename(columns={'Child marriage (%)': 'Female_married_by_15(%)'}, inplace=True)
Marriage.rename(columns={'Unnamed: 2': 'Female_married_by_18(%)'}, inplace=True)
Marriage.rename(columns={'Unnamed: 3': 'Male_married_by_18(%)'}, inplace=True)
Marriage.rename(columns={'Countries and areas': 'Countries'}, inplace=True)
```

```python
Marriage['Female_married_by_18(%)'] = pd.to_numeric(Marriage['Female_married_by_18(%)'], errors='coerce')
Marriage['Female_married_by_15(%)'] = pd.to_numeric(Marriage['Female_married_by_15(%)'], errors='coerce')
Marriage['Male_married_by_18(%)'] = pd.to_numeric(Marriage['Male_married_by_18(%)'], errors='coerce')
```

```python
Marriage=Marriage.drop(Marriage.index[[0,1,2]],axis=0)
```

```python
Marriage.reset_index(drop=True)
Marriage.head()
```

```python
Marriage=Marriage.fillna('')
Marriage.head()
```

```python
for x in Marriage.index:
 if Marriage.loc[x,"Female_married_by_15(%)"]=='':
    Marriage.drop(x,inplace=True)

Marriage.head()
```

```python
for x in Marriage.index:
 if Marriage.loc[x,"Female_married_by_18(%)"]=='':
    Marriage.drop(x,inplace=True)
```

```python
for x in Marriage.index:
 if Marriage.loc[x,"Male_married_by_18(%)"]=='':
    Marriage.drop(x,inplace=True)
```

```python
Marriage.head()
```

```python
Marriage.reset_index(drop=True)
```

```python
Completion=pd.read_excel('https://github.com/shimjide123/datman/raw/main/Completion_rate_2022_formatted%20(1).xlsx')
```

```python
Completion.head()
```

```python
Completion=Completion.fillna('')
Completion.head()
```

```python
for x in Completion.index:
 if Completion.loc[x,"Female_completion_rate"]=='':
    Completion.drop(x,inplace=True)

for x in Completion.index:
 if Completion.loc[x,"Total"]=='':
    Completion.drop(x,inplace=True)

for x in Completion.index:
 if Completion.loc[x,"Male_completion_rate"]=='':
    Completion.drop(x,inplace=True)

for x in Completion.index:
 if Completion.loc[x,"Rural_completion_rate"]=='':
    Completion.drop(x,inplace=True)

for x in Completion.index:
 if Completion.loc[x,"Urban_completion_rate"]=='':
    Completion.drop(x,inplace=True)


Completion.reset_index(drop=True)
```

```python
merged1 = pd.merge(Marriage, Completion, how='inner', on=['Countries'],indicator=True) #outer=keep all
merged1.head(131)
```
